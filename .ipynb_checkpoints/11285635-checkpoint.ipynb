```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
dv = pd.read_csv('red_wine.csv')
dv
```

```python
dv.head(10)
```

```python
plt.figure(figsize=(12, 10))
dv.hist(bins= 20, color= 'blue', edgecolor= 'black', grid=False, linewidth=1.0, xlabelsize= 8, ylabelsize=8)
plt.tight_layout(rect=(0, 0,1.2, 1.2))
plt.show()
plt.savefig('plot5.png')
```

```python
dv.groupby('pH').count().plot()
plt.savefig('plot.png')
```

```python
dv.plot(x= 'residual sugar', y= 'density', xlabel= 'residual sugar', ylabel= 'density', kind= 'scatter')
plt.savefig('plot2.png')
```

```python
dv.plot(x= 'free sulfur dioxide', y= 'total sulfur dioxide', xlabel= 'free sulfur dioxide', ylabel= 'total sulfur dioxide', kind= 'scatter')
plt.savefig('plot3.png')
```

```python
dv.plot(x= 'alcohol', y= 'quality', xlabel= 'alcohol', ylabel= 'quality', kind= 'scatter')
plt.savefig('plot4.png')
```
